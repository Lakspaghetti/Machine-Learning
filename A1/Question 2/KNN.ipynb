```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.colors as cm
```

```python
#loading data
data = np.loadtxt("MNIST-5-6-Subset.txt")
data = np.split(data, 1877)
labels = np.loadtxt("MNIST-5-6-Subset-Labels.txt")
```

```python
#defining training set
train_x = data[0:100]
train_y = labels[0:100]
```

```python
#looking at first picture 
s = 1
np.reshape(train_x[0],(28,28))
plt.imshow(np.reshape(train_x[s],(28,28),order='f'), 'gray', cm.NoNorm())
```

```python
#making validation sets

train_x = data[0:100]
train_y = labels[0:100]
x_val = []
y_val = []
num = [10,20,40,80]

for s in num:
    for i in range(1,6):
        x_val.append(np.array(data[101+(i*s):100+(i+1)*s+1]))
        y_val.append(labels[101+(i*s):100+(i+1)*s+1])     

#calculating distances and finding average loss(empirical loss)
average_loss = []

for n in range(0,20):
    val_loss = []
    for i in range(len(x_val[n])):
        val_dist = np.diag(np.matmul((train_x - x_val[n][i]), np.transpose(train_x - x_val[n][i])))
        sort_val = np.argsort(val_dist)
        prediction = np.round((np.cumsum((np.take_along_axis(train_y, sort_val, 0))[0:50]))/ np.arange(1,51))
        val_loss.append(prediction != y_val[n][i])
    average_loss.append(np.average((val_loss), axis=0))
```

```python
#plotting 4 figures with the 5 validation sets in each, with different n values
s=0
for i in range(20):
    if i%5 == 0:
        plt.figure()
        plt.title(f'Empirical loss as a function of K-neighbours, n={num[s]}')
        s = s+1
    plt.plot(np.arange(1,51),average_loss[i], label=f'i={i+1}')
    plt.xlabel('K-Nearest neighbours')
    plt.ylabel('Empirical loss')
    plt.legend();
    plt.savefig(f'figur {s}')
```
