# Illustration of Markov’s, Chebyshev’s, and Hoeffding’s Inequalities

```python
from itertools import count
import matplotlib.colors as cm
import matplotlib.pyplot as plt
import numpy as np
from math import comb

size = 1000000
n = 20
p = 0.5

reps = np.random.binomial(n, p, size)
PercentualReps = reps/n
alpha = np.arange(0.5,1.05,0.05)
count = [(np.sum(PercentualReps >= a)) for a in alpha]
frequency = np.asarray(count)/size

EX = 0.5
MarkovsRight = EX/(alpha)

VARX = 0.5 * ((0 - 0.5) ** 2) + 0.5 *((1 - 0.5) ** 2)
ChebyshevsRight = VARX/(n * ((alpha - 0.5) ** 2))
ChebyshevsRight[ChebyshevsRight > 1.0] = 1.0

Hoeffdinger = np.exp((-2) * n * ((alpha - 0.5) ** 2))


print(count)
print(alpha)

```

```python
plt.title("plot of empirical frequency over \u03B1")
plt.xlabel('alpha')
plt.ylabel('frequency')
plt.plot(alpha, frequency, color='green', label="Observations")
plt.plot(alpha, MarkovsRight, color='red', label="markov")
plt.plot(alpha, ChebyshevsRight, color='blue', label="Chebyshev")
plt.plot(alpha, Hoeffdinger, color = 'purple', label="hoeffding")
plt.legend(loc=1)
plt.show()
```

```python
print(f"Exact probability for \u03B1=0.95 is: {comb(20, 19) * (0.5 ** 19) * ((1 - 0.5) ** (20-19)) + comb(20, 20) * (0.5 ** 20) * ((1 - 0.5) ** (20-20))}")
print(f"Exact probability for \u03B1=1.0 is: {comb(20, 20) * (0.5 ** 20) * ((1 - 0.5) ** (20-20))}")
```

# Hoeffding corollary 2.5 left hand illustrated

```python
#The following code will give an illustration of why the right hand bound is so low - Note this doesn't represent the bound however!
#Use the Run All button in the top to get different results, dont run this cell alone as this will just give the same results
hoefcount = [(np.sum(PercentualReps - 0.5 >= a)) for a in alpha]
hoefcount2 = [(np.sum(0.5 - PercentualReps >= a)) for a in alpha]


print(hoefcount)
print(hoefcount2)
print(alpha)
```
