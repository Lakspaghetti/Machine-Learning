```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn import datasets
import matplotlib.colors as cm

import matplotlib.pyplot as plt
%matplotlib inline
```

Load handwritten digits data:

```python
digits = datasets.load_digits()
imshape = digits.images[0].shape  # The 2D image shape
X = digits.images.reshape(digits.images.shape[0],-1)  # Image
```

Do the PCA:

Plot the eigenspectrum:

```python
pca = PCA()
pca.fit(X)
eigenvalues = pca.singular_values_**2
explained_variance = eigenvalues / np.sum(eigenvalues)
variance_explained_per_component = pca.explained_variance_ / np.sum(pca.explained_variance_)
cumulative_variance_explained = np.cumsum(variance_explained_per_component)

plt.yscale('log')
plt.xlabel('eigenvalue')
plt.ylabel('eigenvalue number')
plt.plot(explained_variance)
```

Find out if 10 components are enough to explain 80% of the variance:

```python
plt.plot(cumulative_variance_explained)
plt.grid()
plt.xlabel('Number of principal components included')
plt.ylabel('Proportion of variance explained')
plt.title('variance explained as a function of number of PCs included')

print(
    f"variance explained by {10} PCs: {cumulative_variance_explained[9]}")
```

Plot "eigendigits":

```python
eigenvectors = pca.components_
plt.figure(figsize=(12,7))
for i in range(5):
    plt.subplot(1,5,i+1) #rows of subplots, columns of subplots, which subplot
    plt.axis('off')
    plt.title(f'digit {i+1}')
    plt.imshow(eigenvectors[i].reshape(imshape))
```
